## 위의 구성 내용을 한칸에 넣어서 함수화 시키기 

```python
import os, re 

def parsingLangFreq( file_path ) :
    # 언어코드 획득 
    name = os.path.basename( file_path )
    p = re.compile( '^[a-z]{2}' )
    lang_code = p.match(name).group()
    lang_code
    
    # 단어의 빈도 
    with open( file_path, encoding='utf-8' ) as f : 
    # 읽어서 
        text = f.read().lower()  
        p = re.compile('[^a-z]*')
        text = p.sub( '', text )
    
    counts = [ 0 for n in range(26) ]
    ASCII_A = ord('a')
    for ch in text : 
        counts[(ord(ch)-ASCII_A)] += 1
        
    frequences = list( map( checkNorm, counts ) )
    
    # 결과 리턴
    return( lang_code, frequences )

print ( parsingLangFreq( file_list[0] ) )    
```

```python
# 훈련용 데이터, 테스트용 데이터 로드 
# 중간 경로가 train or test 

def load_data( path = 'train' ) : 
    
    # 파일 경로 획득 
    file_lists = glob.glob('./content/{}/*.txt'.format(path) ) # 포맷팅 
    
    # 정답을 담을 리스트 
    labels = list() # 국가
    freqs = list()  # 빈도
    
    # 파일 경로를 한개씩 뽑아서
    for f in file_lists : 
        # 파일 1개당 정답, 빈도수 리스트 리턴 
        lang, freq = parsingLangFreq( f )
        labels.append(lang)
        freqs.append(freq)
    
    return { 'labels' : labels, 'freqs' : freqs }

train_data = load_data()
test_data = load_data('test')
        
```

```python
# 덤프 ( 현 단계 결과물을 저장 )
import json

with open('labels_freqs_data.json', 'w') as f :
    # json 구조로 저장
    json.dump( [train_data, test_data], f )
```
