- 사이트 : https://en.wikipedia.org/wiki/BTS_(band)

```python
target_url = 'https://en.wikipedia.org/wiki/BTS_(band)'
target_url
```

```python
import urllib.request as req 
from bs4 import BeautifulSoup 
```

```python
res = req.urlopen(target_url)
soup = BeautifulSoup(res, 'html5lib') ## Beautiful 모듈을 쓰는 방법 
```

```python
#type(soup),soup
```

```python
# 대상 태그들 모으기
ps = soup.select('#mw-content-text p') ##태그(p)이름을 맨 뒤, 클래스 이름(mw~) 앞 # 붙이는 게 BeautifulSoup방식
len(ps), type(ps)
```

```python
# ps에서 한 개씩 뽑아서 텍스트 출력 

for p in ps :
    print('[%s]' % p.text.strip() )
    break
```

```python
# ps에서 79개를 다 뽑아서 tmp 리스트에 담기

tmp = list()
for p in ps :
    tmp.append('[%s]' % p.text.strip() )
    
tmp
```

```python
# 목적 : 텍스트 여러 개를 리스트에 모은다
# 리스트 내포를 이용하여 79개의 p태그에서 텍스트 추출하여 리스트에 구성원으로 추가하기

tmp = [ p.text.strip() for p in ps ] ## 위의 for문을 한줄식으로!!
tmp[:10]
```

```python
# 리스트에 존재하는 문자열들을 한 개의 문자열로 통합하기

src_txt = ' '.join(tmp)
len(scr_txt), scr_txt[:100]
```

```python
## 가독성 올리기 
# 영어는 소문자로 통일(lower)
# 영어를 제외하고 모두 제거(정규식)
# 띄어쓰기 기준으로 공백 남기기(정규식 중에 공백은 두자)

import re 

p = re.compile('[^a-zA-Z ]*') 
## Z 뒤에 한 칸 띄움으로써 공백 만듬(제일 쉬운 방법, 조금 안 맞을 수 있음 )
##re.compile() : 정규식 패턴을 파이썬이 사용할 수 있는 정규식 객체로 컴파일 해주는 기능을 한다. re.compile()을 통해 객체로 변환된 정규식은 match()와 search()를 통해 검색할 때 패턴으로 이용할 수 있다.
tmp = p.sub( '', src_txt ) ## sub() : 대체? p가 아닌 것을 공백으로 대체한다 !!! 
tmp[:100]
```
